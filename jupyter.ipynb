```python
import numpy as np
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(8,3))
ax = fig.add_axes([0, 0, 1, 1])
x = np.linspace(0, 5, 10)

ax.plot(x, x**2,'b.-', label="x**2")
ax.plot(x,x**3,'g-.',label='X**3')
plt.legend(loc=10)
plt.show()
```

```python
fig,ax=plt.subplots(1,3,figsize=(12,4))
ax[0].plot(x,x**2,x,x**3)
ax[0].set_title("first")

ax[1].plot(x,x**2,x,x**3)
ax[1].axis('tight')
ax[1].set_title("second")

ax[2].plot(x,x**2,x,x**3)
ax[2].set_xlim([2,5])
ax[2].set_ylim([0,60])
plt.show()
```

```python
import numpy as np
x=np.arange(0,100)
y=x*2
z=x**2
```

```python
import matplotlib.pyplot as plt
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,y,'Black',label='x axis')
ax.set_title('SNEKAN',color='r')
plt.legend(loc=3)
plt.show()
```

```python
fig = plt.figure()
ax1 = fig.add_axes([0, 0, 1, 1])
ax1.plot(x,z,'r')
ax1.set_xlim([0,40])
ax1.set_ylim([0,2000])
ax2=fig.add_axes([0.2,0.5,.4,.4])
ax2.plot(x,y,'g')
ax2.set_xlim([20,22])
ax2.set_ylim([30,50])
```
